```python
import seaborn as sns
import plotly.express as px
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic
```

```python
px.scatter(titanic , x="age",y="fare")  # plotly express scatter function
```

```python
# Here we're using Plotly Express's scatter() function to create a scatter plot. The x parameter specifies the data for the x-axis ("age" column), and the y parameter specifies the data for the y-axis ("fare" column).
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
px.box(tips,x='day',y='tip')  # craeting a box plot using plotly express
```

```python
# Here we're using the tips dataset provided by Plotly Express . The box() function from Plotly Express is used to create a box plot. The x parameter specifies the data for the x-axis ("day" column), and the y parameter specifies the data for the y-axis ("tip" column).
```

```python
# Q.3
```

```python
px.histogram(tips, x="sex",y="total_bill",pattern_shape="smoker",color="day")
```

```python
# Here histogram() function from Plotly Express is used to create a histogram. The x parameter specifies the data for the x-axis ("sex" column), the y parameter specifies the data for the y-axis ("total_bill" column), the color parameter uses the "day" column to differentiate colors, and the pattern_shape parameter uses the "smoker" column to differentiate pattern shapes
```

```python
# Q.5 The distplot() function was a versatile visualization function provided by Seaborn for creating histograms and kernel density plots (KDE) of univariate data. It combined both a histogram and a KDE plot to provide insights into the distribution of a single continuous variable.
#     Plotly Express does not provide a direct distplot() function so we can use another library here.
```

```python
import numpy as np
import plotly.figure_factory as ff
hist_data=[np.random.rand(100)]
labels=['distplot']
figure=ff.create_distplot(hist_data,labels)
figure.show()
```
